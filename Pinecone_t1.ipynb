```python
#index 생성 (Table)
pc.create_index(
    name="quicktest",
    dimension=384, # Replace with your model dimensions
    metric="cosine", # Replace with your model metric
    spec=ServerlessSpec(
        cloud="aws",
        region="us-east-1"
    )
)
```

```python
#index 불러오기
index = pc.Index('quicktest')
print(index)
```

```python
#로컬에서 json 파일 로드
from google.colab import files
uploaded = files.upload()

import json

file_name = list(uploaded.keys())[0]
with open(file_name, 'r') as file:
    data = json.load(file)

print(data)
```

```python
# 각 항목을 하나씩 처리합니다.
for item in data:
    print(item)
    id = str(item['recruitId'])
    value = embeddings.embed_query(item['content'])
    #del item['content']
    print(id, value)
    index.upsert(
        vectors=[
            {"id": id, "values": value, "metadata" : item}
        ]
    )
```

```python
#list record ids
for ids in index.list():
    print(ids)
```

```python
#Index 확인
index.describe_index_stats()
```

```python
query = "AI를 이용한 빅데이터 분석 직무"
query_embedding = embeddings.embed_query(query)
#print(query_embedding)

#유사도 측정 (Query)
res = index.query(
    #namespace="ns1",
    #filter={"year": 1979},
    vector=query_embedding,
    top_k=5,
    include_values=True,
    include_metadata=True
)

# get list of retrieved text
contexts = [item['metadata']['content'] for item in res['matches']]

augmented_query = "\n\n---\n\n".join(contexts)+"\n\n-----\n\n"+query

print(augmented_query)
```

```python
#로컬에서 json 파일 로드
from google.colab import files
uploaded = files.upload()

import json

file_name = list(uploaded.keys())[0]
with open(file_name, 'r') as file:
    member_data = json.load(file)

print(member_data)
```

```python
##https://docs.pinecone.io/guides/data/filter-with-metadata

filter_conditions = {
    "$and": [
        {"major": {"$in": member_data['major']}},
        {"$or": [
            {"comp": {"$in": member_data['comp']}},
            {"aragp": {"$in": member_data['aragp']}},
            {"job": {"$in": member_data['job']}},
            {"dgr": {"$in": member_data['dgr']}}
            ]
        }
        ]
    }

filter_conditions
```

```python
query = "AI를 이용한 빅데이터 분석 직무"
query_embedding = embeddings.embed_query(query)
#print(query_embedding)

#유사도 측정 (Query)
res = index.query(
    #namespace="ns1",
    filter=filter_conditions,
    vector=query_embedding,
    top_k=5,
    include_values=True,
    include_metadata=True
)

# get list of retrieved text
contexts = [item['metadata']['content'] for item in res['matches']]

augmented_query = "\n\n---\n\n".join(contexts)+"\n\n-----\n\n"+query

print(augmented_query)
```
